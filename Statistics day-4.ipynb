```python
import numpy.random as np
import seaborn as sns
import matplotlib.pyplot as plt
population_size = 1000000 # N i.e size of population.
population = np.rand(1000000) # population data. Randomply it will will generate 1000000 data from this range(0,1).

```

```python
population # Population data
```

```python
number_of_samples = 10000 # m i.e number of samples
sample_means = np.rand(number_of_samples) # randomly picking up 10000 samples which is in the (0,1) range.
sample_size = 35 # n i.e  sample size.
```

```python
sample_means 
```

```python
import numpy
c = np.rand(number_of_samples) # Randomly selecting 10000 i.e m value i.e no. of samples data from (0,1) range.
for i in range(0, number_of_samples):
    c = np.randint(1, population_size, sample_size) # Randomly It selects the sample data from population data of sample size 35.
    sample_means[i] = numpy.mean(c) # It calculates the mean of each sample data.
```

```python
np.randint(1, population_size, sample_size)
```

```python
sample_means
```

```python
sample_means.shape # Size of sample mean is 10000.
```

```python
import matplotlib.pyplot as plt
plt.subplot(1,2,1)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
sns.distplot(sample_means,bins=int(180/5),hist = True,kde = False)
plt.title('Histogram of Sample mean',fontsize=20)
plt.xlabel('Sample mean',fontsize=20)
plt.ylabel('Count',fontsize=20)
plt.subplot(1,2,2)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
sns.distplot(sample_means,hist = False,kde = True)
plt.title('Density of Sample mean',fontsize=20)
plt.xlabel('Sample mean',fontsize=20)
plt.ylabel('Density',fontsize=20)
plt.subplots_adjust(bottom=0.1, right=2, top=0.9)
```
